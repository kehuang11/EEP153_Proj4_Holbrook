```python
from eep153_tools.sheets import read_sheets
import numpy as np
import pandas as pd

def get_clean_sheet(key,sheet=None):

    df = read_sheets(key,sheet=sheet)
    return get_clean_df(df)

def get_clean_df(df):
    df.columns = [c.strip() for c in df.columns.tolist()]

    df = df.loc[:,~df.columns.duplicated(keep='first')]

    df = df.drop([col for col in df.columns if col.startswith('Unnamed')], axis=1)

    df = df.loc[~df.index.duplicated(), :]
    
    return df
```

```python
#Read Expenditures Sheets from Google Sheets
lsms = get_clean_sheet(InputFiles["Expenditures"][0], sheet = InputFiles["Expenditures"][1])
```

```python
def unionize_names(lsms):
    lsms = lsms.replace(to_replace= "Beef", value = "Beef including minced sausage")
    lsms = lsms.replace(to_replace="Cooking oil", value = "Cooking oil (vegetable oil)")
    lsms = lsms.replace(to_replace= "Dried/salted/canned fish", value = "Dried/salted/canned fish and seafood (including dagaa)")
    lsms =lsms.replace(to_replace= "Spinach, cabbage, other green vegetables", value ="Spinach, cooked (/cabbage and other green vegetables)")
    lsms =lsms.replace(to_replace= "Tea (dry)", value ="Tea dry")
    lsms =lsms.replace(to_replace= "Onions, tomatoes, carrots, green pepper, other viungo", value ="Onion, raw (/tomatoes, carrots and green pepper, other viungo)")
    lsms =lsms.replace(to_replace= "Rice (husked)", value ="Rice (Husked) (white grain raw)")
    lsms =lsms.replace(to_replace= "Rice (paddy)", value ="Rice (Paddy)")
    lsms =lsms.replace(to_replace= "Maize (flour)", value ="Maize (Flour, dry)")
    lsms =lsms.replace(to_replace= "Maize (green, cob)", value ="Maize (Green, Cob, immature)")
    lsms =lsms.replace(to_replace= "Maize (grain)", value ="Maize (Grain, dry, raw)")
    lsms =lsms.replace(to_replace= "Millet and sorghum (grain)", value ="Millet and Sorghum (Grain)")
    lsms =lsms.replace(to_replace= "Millet and sorghum (flour)", value ="Millet and Sorghum (Flour)")
    lsms =lsms.replace(to_replace= "Bread", value ="Bread, white")
    lsms =lsms.replace(to_replace= "Pasta", value ="Macaroni (/Spaghetti)")
    lsms =lsms.replace(to_replace= "Cassava fresh", value ="Cassava Fresh, raw")
    lsms =lsms.replace(to_replace= "Sweet potatoes", value ="Sweet potatoes, fresh AP")
    lsms =lsms.replace(to_replace= "Yams/cocoyams", value ="Yams/cocoyams raw")
    lsms =lsms.replace(to_replace= "Irish potatoes", value ="Irish potatoes, cooked (English)")
    lsms =lsms.replace(to_replace= "Plantains", value ="Cooking bananas (/Plantains)")
    lsms =lsms.replace(to_replace= "Other starches", value ="Other starches (edible)")
    lsms =lsms.replace(to_replace= "Sweets", value ="Sweets (Candy, Chocolate)")
    lsms =lsms.replace(to_replace= "Honey, syrups, jams, marmalade, jellies, canned fruits", value ="Honey (/Syrups, jams, marmalade, jellies, canned fruits)")
    lsms =lsms.replace(to_replace= "Peas, beans, lentils and other pulses", value ="Lentil, cooked / Peas (/Beans, lentils and other pulses)")
    lsms =lsms.replace(to_replace= "Groundnuts", value ="Groundnuts in shell/shelled")
    lsms =lsms.replace(to_replace= "Coconuts", value ="Coconuts (mature/immature)")
    lsms =lsms.replace(to_replace= "Seeds and nut products", value ="Sunflower seed / Seeds and products from nuts/seeds (excl. cooking oil)")
    lsms =lsms.replace(to_replace= "Ripe bananas", value ="Ripe Bananas")
    lsms =lsms.replace(to_replace= "Citrus fruits", value ="Oranges (citrus fruits)")
    lsms =lsms.replace(to_replace= "Pork", value ="Pork including sausages and bacon")
    lsms =lsms.replace(to_replace= "Chicken", value ="Chicken and other poultry")
    lsms =lsms.replace(to_replace= "Cricket", value ="Cricket (Other domestic/wild meat products)")
    lsms =lsms.replace(to_replace= "Fresh fish", value ="Raw Fresh fish and seafood (including dagaa)")
    lsms =lsms.replace(to_replace= "Butter", value ="Butter refined - ghee")
    lsms =lsms.replace(to_replace= "Wheat (flour)", value ="Wheat flour (whole grain)")
    lsms =lsms.replace(to_replace= "WHEAT, BARLEY, GRAIN, AND OTHER CEREALS", value ="Other cereal products")
    lsms =lsms.replace(to_replace= "Soft drinks", value ="Mango juice (soda, juice, water)")
    return lsms
```
